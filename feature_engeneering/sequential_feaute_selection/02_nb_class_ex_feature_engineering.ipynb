```python
"""
Feature Engineering Example: Sequential Feature Selection

How to perform feature selection using
Forward Selection and Backward Selection with scikit-learn.

Dataset: Titanic
Model used: Decision Tree Classifier
"""

import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SequentialFeatureSelector

```

```python
# Load Dataset

titanic = pd.read_csv("/content/sample_data/titanic.csv")

# Standardize column names (lowercase and replace "/" with "_")
column_names = {col: col.lower().replace("/", "_") for col in titanic.columns}
titanic.rename(columns=column_names, inplace=True)

print("Dataset preview:")
print(titanic.head())
```

```python
# Data Preparation

# Convert the 'sex' variable from categorical to numeric
# male = 0, female = 1
titanic["sex"] = titanic["sex"].map({"male": 0, "female": 1})

# Define features and target variable
# Drop non-numeric columns: 'name' (already dropped), 'ticket', 'cabin', 'embarked'
# 'passengerid' is also typically dropped as it's an identifier, not a feature
X = titanic.drop(['survived', 'name', 'ticket', 'cabin', 'embarked', 'passengerid'], axis=1)
y = titanic['survived']
```

```python
# Model Definition


# Decision Tree classifier used for feature selection
dt = DecisionTreeClassifier(
    max_depth=5,
    min_samples_leaf=5,
    random_state=111
)
```

```python
# Backward Feature Selection

"""
Backward Selection:
Starts with all features and progressively removes the least useful ones
until reaching the desired number of features.
"""

selector_backward = SequentialFeatureSelector(
    estimator=dt,
    n_features_to_select=4,
    direction="backward",
    scoring="precision",
    cv=5
)

selector_backward.fit(X, y)

# Selected features
backward_results = pd.DataFrame({
    "feature": X.columns,
    "selected": selector_backward.support_
})

print("\nBackward Selection Results:")
print(backward_results)
```

```python
X = titanic.drop(['survived', 'name'], axis=1)
X.head()
```

```python
y = titanic.survived
```

```python
dt = DecisionTreeClassifier(max_depth=5, min_samples_leaf=5, random_state=111)
```

```python
# estimator: qué estimador utilizar, en este caso un árbol de clasificación
# n_features_to_select: con cuántas variables nos queremos quedar
# direction: backward o forward
# scoring: qué métrica de desempeño utilizar para seleccionar
selector_backward = SequentialFeatureSelector(estimator=dt, n_features_to_select=4, direction="backward",
                                     scoring='precision', cv=5)
selector_backward.fit(X,y)
```

```python
selector_backward.support_
```

```python
pd.DataFrame({'column': X.columns.values, 'selected': selector_backward.support_})
```
